# **Data Loading**

```python
# Importing libraries

import pandas as pd
```

```python
# Loading dataset

df = pd.read_csv('dataset.csv')
```

```python
# Dataset view

df.head()
```

# **Data Cleaning**

```python
# Inspecting missing and duplicate values

print(df.isnull().sum())
print(df.duplicated().sum())
```

# **Data Transformation**

```python
# Dataset columns

df.columns
```

```python
# Clean Payment of minimum amount

df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].replace({'NM' : 'Not Mentioned'})
df
```

```python
# Handling ambiguous type of loan data

df['Type_of_Loan'] = df['Type_of_Loan'].replace({'No Data': 'Not Specified', None: 'Not Specified'})
```

```python
# Split loan types into lists

df['Loan_Types_List'] = df['Type_of_Loan'].str.split(', ')
df.head()
```

```python
# Flatten the list of all loans into one long list, strip whitespace, and get unique loan types

all_loans = df['Loan_Types_List'].explode().str.replace('and ', '', regex=False).str.strip().unique()
all_loans
```

```python
# For each unique loan type, creating a new column (binary)

for loan in all_loans:
    df[loan] = df['Loan_Types_List'].apply(lambda x: 1 if loan in x else 0)
df
```

```python
# Transforming to csv file

df.to_csv('cleaned_dataset.csv', index=False)
```
